## 2. Redução de Dimensionalidade

Os algoritmos de aprendizado de máquina dependem de funções de otimização para aprender com os exemplos fornecidos. No entanto, otimizar esses algoritmos pode ser desafiador quando lidamos com conjuntos de dados de alta dimensionalidade, ou seja, muitas variáveis.

Uma técnica comum para lidar com esse problema é o método de Componentes Principais, conhecido como Principal Component Analysis (PCA), que é amplamente utilizado para redução de dimensionalidade.

### Exemplo de Redução de Dimensionalidade utilizando PCA

```python
# Importação das bibliotecas necessárias
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Carrega o conjunto de dados 'iris.csv' em um DataFrame chamado df_iris
df_iris = pd.read_csv('iris.csv')

# Exibe as primeiras linhas do DataFrame df_iris para visualização inicial
df_iris.head()
```

```python
# Separação dos atributos das classes
X = df_iris.iloc[:, :-1]  # Atributos
y = df_iris.iloc[:, -1]   # Classes

# Instanciação do modelo PCA com n_dimensions
n_dimensions = 2
pca = PCA(n_dimensions)

# Aplicação do PCA aos dados
X_pca = pca.fit_transform(X)

# Visualização dos resultados
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('Dimensão 1')
plt.ylabel('Dimensão 2')
plt.title(f'Redução de dimensionalidade usando PCA (Gráfico com 2 dimensões)')
plt.colorbar(label='Classe')
plt.show()

# Percentual de variância explicada por cada componente principal
explained_variance_ratio_percent = pca.explained_variance_ratio_ * 100
print("Percentual de variância explicada por cada dimensão:")
for i, explained_variance in enumerate(explained_variance_ratio_percent):
    print(f"Dimensão {i+1}: {explained_variance:.2f}%")
```
